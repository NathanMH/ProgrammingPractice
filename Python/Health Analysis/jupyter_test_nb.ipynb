```python
import random
import matplotlib.pyplot as plt
import numpy as np


def sieve():
    MAX_PRIME = 10000
    count = 0

    sieve = [True] * MAX_PRIME
    for i in range(2, MAX_PRIME):
        if sieve[i]:
            count += 1
            for j in range(i * i, MAX_PRIME, i):
                sieve[j] = False

    print("Total Primes: " + str(count))
    
sieve()
```

```python
import random
import random
import matplotlib.pyplot as plt
import numpy as np

def coin_flip_odds():
    counts = {"1": 0}

    flips = ""
    for i in range(1, 100):
        flip = random.randint(0, 1)
        flips += str(flip)
    print(flips)

    # for j in range(0, len(flips) - 2):
    j = 0
    while j < (len(flips) - 2):

        k = j + 1

        if flips[j] != flips[k]:
            counts["1"] += 1
            j += 1

        else:
            substr_len = 1

            while flips[k] == flips[j]:
                substr_len += 1
                k += 1
                j += 1

            if str(substr_len) in counts:
                counts[str(substr_len)] += 1
            else:
                counts[str(substr_len)] = 1

    print(sorted(counts.items()))

    x_vals = counts.keys()
    y_vals = counts.values()
    y_pos = np.arange(len(counts.keys()))
    plt.bar(y_pos, y_vals, align="center")
    plt.xticks(y_pos, x_vals)
    plt.ylabel("total count")
    plt.title("count of 1s and 0s")
    plt.show()

coin_flip_odds()
```
